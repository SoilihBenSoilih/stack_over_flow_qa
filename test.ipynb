```python
from stackapi import StackAPI
import time

def fetch_questions_with_accepted_answers(tag, max_questions=1000):
    SITE = StackAPI('stackoverflow')
    SITE.page_size = 100
    SITE.max_pages = 5  # Pour atteindre jusqu'à 1000 questions

    questions_with_accepted = []
    accepted_answer_ids = []

    for page in range(1, SITE.max_pages + 1):
        print(f"Fetching page {page} for tag '{tag}'...")
        questions = SITE.fetch('questions', page=page, sort='creation', order='desc', tagged=tag, filter='default')
        for question in questions['items']:
            if 'accepted_answer_id' in question:
                questions_with_accepted.append(question)
                accepted_answer_ids.append(question['accepted_answer_id'])
        time.sleep(1)  # Respecter les limites de l'API

    return questions_with_accepted, accepted_answer_ids

def fetch_accepted_answers(answer_ids):
    SITE = StackAPI('stackoverflow')
    answers = []
    for i in range(0, len(answer_ids), 100):
        batch_ids = answer_ids[i:i + 100]
        print(f"Fetching answers for IDs {batch_ids}...")
        response = SITE.fetch('answers/{ids}', ids=batch_ids, filter='withbody')
        answers.extend(response['items'])
        time.sleep(1)  # Respecter les limites de l'API
    return answers

# Tags à traiter
tags = ['python', 'java']
all_questions = []
all_answer_ids = []

for tag in tags:
    questions, answer_ids = fetch_questions_with_accepted_answers(tag)
    all_questions.extend(questions)
    all_answer_ids.extend(answer_ids)

# Supprimer les doublons
unique_answer_ids = list(set(all_answer_ids))

# Récupérer les réponses acceptées
accepted_answers = fetch_accepted_answers(unique_answer_ids)

# Afficher les résultats
for question in all_questions:
    print(f"Question: {question['title']}")

```

```python
# Créer un dictionnaire des réponses acceptées
answer_dict = {answer['answer_id']: answer for answer in accepted_answers}

# Afficher les questions avec leurs réponses acceptées
for question in all_questions:
    accepted_id = question.get('accepted_answer_id')
    accepted_answer = answer_dict.get(accepted_id)
    if accepted_answer:
        print(f"Question: {question['title']}")
        print(f"Réponse acceptée: {accepted_answer['body'][:200]}...\n")
```
